```python
import matplotlib.pyplot as plt
# Teslaの株価データを取得
tesla_data = yf.download("TSLA", start="2020-01-01", end="2023-01-01")

# グラフを作成する関数
def make_graph(stock_data, title="Stock Price Over Time"):
    plt.figure(figsize=(12, 6))
    plt.plot(stock_data.index, stock_data['Close'], label='Close Price')  # 終値をプロット
    plt.title(title)
    plt.xlabel("Date")
    plt.ylabel("Stock Price (USD)")
    plt.legend()
    plt.show()

# Teslaの株価グラフを作成
make_graph(tesla_data, title="Tesla Stock Price Over Time")
```
